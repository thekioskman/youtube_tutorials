```python
import plotly.express as px
import pandas as pd


data = pd.read_csv("test.csv")
fig = px.scatter(data, x="x" , y="y")
fig.show()
```

```python

data_list = data.values.tolist()
x_list = []
y_list = []
for point in data_list:
    x_list.append(point[0])
    y_list.append(point[1])

#so now data is a list of points in the form [x,y]
```

```python
import plotly.graph_objects as go
#consider now we have a line in the form y = mx+b 
#Note: the max x value is 100
m = 0
b = 0

fig = go.Figure()
fig.add_trace(go.Scatter(x=x_list, y = y_list , mode="markers"))
fig.add_trace(go.Scatter(x=[0,100], y=[b , m*100+b] , mode="lines"))

fig.show()
```

```python
#Create the error function
def get_error(m,b):
    error = 0
    for points in data_list:
        error += ((points[1]- points[0]*m +b)**2)/len(data_list)

    return error
```

```python
#do gradient descent
def get_gradients(m,b):
    partial_m = 0
    partial_b = 0
    for points in data_list:
        partial_m += ((points[1] - m*points[0]+b)*points[0] )/len(data_list)
        partial_b += ((points[1] - m*points[0]+b))/len(data_list)
    
    return -1*partial_m , -1*partial_b
```

```python
#HERE IS A GOOD TIME TO SHOW THE IMPORTANCE OF THE LEARNING RATE
m = 0
b = 0
error = []
alpha = 0.0001
for x in range(1000):
    partial_m , partial_b = get_gradients(m,b)
    m -= partial_m * alpha
    b -= partial_b * alpha
    if x %100 == 0:
        print(x, "m:" ,m, "b",b , "error" , get_error(m,b))

fig = go.Figure()
fig.add_trace(go.Scatter(x=x_list, y = y_list , mode="markers"))
fig.add_trace(go.Scatter(x=[0,100], y=[b , m*100+b] , mode="lines"))

fig.show()

```

```python
error = []
temp_m = []

for m in range(-100,100):
    error.append(get_error(m,0))
    temp_m.append(m)

error_fig = go.Figure()
error_fig.add_trace(go.Scatter(x=temp_m,y=error , mode="markers"))
error_fig.show()
```

```python
#Lets see what is actually happending when our learning rate is too high
#We will see that when the change the learning rate from 0.0001 to 0.001 something crazy will happen
error = []
temp_m = []
m= 0 
b = 0

alpha = 0.001
for x in range(10):
    partial_m , partial_b = get_gradients(m,b)
    m -= partial_m * alpha
    b -= partial_b * alpha
    error_val = get_error(m,b)
    error.append(error_val)
    temp_m.append(m)

print(temp_m)

error_fig = go.Figure()
error_fig.add_trace(go.Scatter(x=temp_m,y=error , mode="markers"))
error_fig.show()
```
